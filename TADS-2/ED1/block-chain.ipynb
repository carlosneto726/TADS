```python
from datetime import datetime
from hashlib import sha256


class ListaCadeia: # Lista Encadeada Cujo o ID seja o HASH
    def __init__(self):
        self.proximo_a_entrar = None
        self.proximo_a_sair   = None
        self.items = {}
    def inserir(self, id, valor):
        # Inserinho os Valores na lista encadeadas
        self.proximo_a_entrar = id
        self.items[self.proximo_a_entrar] = valor

id = ListaCadeia()

class Block():
    def __init__(self, transactions, data, previous_hash):
        self.transactions = transactions
        self.data = data
        self.previous_hash = previous_hash
        self.nonce = 0
        self.timestamp = datetime.now()
        self.hash = self.generate_hash()

    def print_content(self):
        print("Timestamp: ", self.timestamp)
        print("Transactions: ", self.transactions)
        print("Data: ", self.data)
        print("Current Hash: ", self.generate_hash())
        print("Previous Hash: ", self.previous_hash)
        print("\n")

    def generate_hash(self):
        block_contents = str(self.timestamp) + str(self.transactions) + str(self.data) + str(self.previous_hash) + str(self.nonce)
        block_hash = sha256(block_contents.encode())
        return block_hash.hexdigest()


class Blockchain():
    
    def __init__(self):
        self.chain = []
        self.all_data = []
        self.genesis_block()
    
    def genesis_block(self):
        transactions = None
        data = []
        previous_hash = "0"
        self.chain.append(Block(transactions, data, previous_hash))

    def print_blocks(self):
        for i in range(len(self.chain)):
            current_block = self.chain[i]
            print("Block {} {}".format(i, current_block))
            current_block.print_content()  

    def add_block(self, data):
      # Auto Validação para as regras de cada bloco
        if self.validate_chain():
            previous_block_hash = self.chain[len(self.chain)-1].hash
            transactions = data['tipo_transacao']
            new_block = Block(transactions, data, previous_block_hash)
            self.chain.append(new_block)  
            datain = {data['tipo_transacao']: data}
            # Inserindo o Hash com as transações na lista encadeada
            id.inserir(self.chain[len(self.chain)-1].hash, datain)
        

    def validate_chain(self):
        i=1
    
        for i in range(1, len(self.chain)-1):
            current = self.chain[i]
            previous = self.chain[i-1]
            if current.hash != current.generate_hash():
                print("O Hash do bloco atual não é igual ao Hash do bloco  gerado.")
                return False
            if previous.hash != previous.generate_hash():
                print("O Hash do bloco anterior não é igual ao Valor do Hash anterior do bloco atual.")
                return False
            # Regras do négocio
            if 'Nascimento' in id.items[self.chain[i].hash]:
                if 'data_obito' in id.items[self.chain[i].hash]['Nascimento']:
                    return False
                    
            if 'Casamento' in id.items[self.chain[i].hash]:
                if 'conjuge1' in id.items[self.chain[i].hash]['Casamento'] or 'conjuge2' in id.items[self.chain[i].hash]['Casamento']:
                    if 'data_obito' in id.items[self.chain[i].hash]['Casamento']['conjuge1'] or 'data_obito' in id.items[self.chain[i].hash]['Casamento']['conjuge2']:
                        return False

            if 'Separacao' in id.items[self.chain[i].hash]:
                for sep in range(1, len(self.chain)):
                    if 'Separacao' in id.items[self.chain[sep].hash]:
                        if self.chain[sep].hash != self.chain[i].hash:
                            return False
                    if 'Casamento' in id.items[self.chain[sep].hash]:
                        if id.items[self.chain[sep].hash]['Casamento']['conjuge1'] != id.items[self.chain[i].hash]['Separacao']['conjuge1']:
                            return False
                if 'conjuge1' in id.items[self.chain[i].hash]['Separacao'] or 'conjuge2' in id.items[self.chain[i].hash]['Separacao']:
                    if 'data_obito' in id.items[self.chain[i].hash]['Separacao']['conjuge1'] or 'data_obito' in id.items[self.chain[i].hash]['Separacao']['conjuge2']:
                        return False
               
            if 'Viuvez' in id.items[self.chain[i].hash]:
                if 'conjuge1' in id.items[self.chain[i].hash]['Viuvez'] or 'conjuge2' in id.items[self.chain[i].hash]['Viuvez']:
                    if 'data_obito' not in id.items[self.chain[i].hash]['Viuvez']['conjuge1'] or 'data_obito' not in id.items[self.chain[i].hash]['Viuvez']['conjuge2']: 
                        return False
        
        return True
```
